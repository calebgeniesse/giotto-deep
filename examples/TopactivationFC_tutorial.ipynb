```python
from gdeep.topactivation import TopactivationFC as TFC
from gdeep.pipeline import Pipeline
from torch.utils.tensorboard import SummaryWriter
from gdeep.data import TorchDataLoader
from gdeep.models import FFNet
from torch import nn
```

```python
writer = SummaryWriter()
dl = TorchDataLoader(name="MNIST")
dl_tr, dl_ts = dl.build_dataloader(batch_size=32)
```

```python
arch = [28*28, 100, 100, 10]
model = nn.Sequential(nn.Flatten(), FFNet(0, arch= arch ))
loss_fn = nn.CrossEntropyLoss()
pipe = Pipeline(model, (dl_tr, dl_ts), loss_fn, writer)
```

```python
topactiv = TFC(pipe,arch)
```

```python
x = next(iter(dl_tr))[0]
diagrams_training = topactiv.diagrams_accross_training(x,15)
```

```python
diagrams_reversed = [topactiv.reverse_diagrams(diagrams) for diagrams in diagrams_training]
```

```python
from gudhi.representations.vector_methods import Entropy as Entr
import numpy as np
entr = Entr()
entropies = [entr.transform(diagrams) for diagrams in diagrams_reversed]
```

```python
import plotly.express as px
fig = px.line(entropies)
fig.show()
```

```python
data, target = next(iter(dl_tr))
perturbed_image = topactiv.fgsm_attack(data,target,0.05)
```
